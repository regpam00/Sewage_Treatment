```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("Physical Nutrients Microbes (Cleaned).xlsx")
#print(df.head(10))
```

### Coupeville

```python
# Selecting PC samples

coupeville = df.loc[df['Site Name'] == 'CV']
print(coupeville.head(5))
```

Nutrients

```python
# Mean values of nutrients

nutrients_mean_coupeville = coupeville.groupby(["Date"]).mean(numeric_only = True)
print(nutrients_mean_coupeville.loc[:,["Ammonium (NH4+) µM", "Nitrite (NO2-) µM", "Nitrate (NO3-) µM", "Phosphate (PO43-) µM", 
"Silicate (Orthosilicic Acid) Si(OH)4 µM"]])
```

```python
# Standard deviations of nutrients

nutrients_std_coupeville = coupeville.groupby(["Date"]).std(numeric_only = True)
print(nutrients_std_coupeville.loc[:,["Ammonium (NH4+) µM", "Nitrite (NO2-) µM", "Nitrate (NO3-) µM", "Phosphate (PO43-) µM", 
"Silicate (Orthosilicic Acid) Si(OH)4 µM"]])
```

```python
# Variance of nutrients

nutrients_var_coupeville = coupeville.groupby(["Date"]).var(numeric_only = True)
print(nutrients_var_coupeville.loc[:,["Ammonium (NH4+) µM", "Nitrite (NO2-) µM", "Nitrate (NO3-) µM", "Phosphate (PO43-) µM", 
"Silicate (Orthosilicic Acid) Si(OH)4 µM"]])
```

Graphs

```python
old_dates = [date for date, df in df.groupby("Date")]
new_dates = list(pd.to_datetime(old_dates).strftime('%m/%d/%Y'))
print(new_dates)
```

```python
# Mean values of nutrients

nutrients_mean_coupeville = coupeville.groupby(["Date"]).mean(numeric_only = True)
dates = ['08/24/2010', '08/25/2010', '08/26/2010', '11/21/2010', '11/23/2010', '02/15/2011', '02/16/2011', '02/17/2011', '05/17/2011', '05/18/2011', '05/19/2011']

nutrients_mean_coupeville_1 = pd.DataFrame({
    "Dates" : dates,
    "Ammonium (NH4+)" : list(nutrients_mean_coupeville.loc[:,"Ammonium (NH4+) µM"]),
    "Nitrite (NO2-)" : list(nutrients_mean_coupeville.loc[:,"Nitrite (NO2-) µM"]),
    "Nitrate (NO3-)" : list(nutrients_mean_coupeville.loc[:,"Nitrate (NO3-) µM"]),
    "Phosphate (PO43-)" : list(nutrients_mean_coupeville.loc[:,"Phosphate (PO43-) µM"]),
    "Silicate (Si(OH)4)" : list(nutrients_mean_coupeville.loc[:,"Silicate (Orthosilicic Acid) Si(OH)4 µM"])
})
nutrients_mean_coupeville_2 = pd.melt(nutrients_mean_coupeville_1, id_vars = "Dates", var_name = "Nutrients", value_name = "Concentration (micromoles per liter)")
c = sns.catplot(data = nutrients_mean_coupeville_2, x = "Dates", y = "Concentration (micromoles per liter)", hue = "Nutrients", kind = "bar")
c.set_xticklabels(rotation = 90)
plt.title("Average Nutrient Concentrations Collected Per Sample Day in Coupeville", loc = "center")
```

```python
# Mean values of microbes

microbes_mean_coupeville = coupeville.groupby(["Date"]).mean(numeric_only = True)
dates = ['08/24/2010', '08/25/2010', '08/26/2010', '11/21/2010', '11/23/2010', '02/15/2011', '02/16/2011', '02/17/2011', '05/17/2011', '05/18/2011', '05/19/2011']
microbes =['Nano PP','Eukaryotic PP', 'HNA Bacteria', 'LNA Bacteria', 'Synechococcus', 'Total Bacteria']

microbes_mean_coupeville_1 = pd.DataFrame({
    "Dates" : dates,
    "Nanophytoplankton" : list(microbes_mean_coupeville.loc[:,"Nanophytoplankton (cells/ml)"]),
    "Eukaryotic Phytoplankton" : list(microbes_mean_coupeville.loc[:,"FACS Eukaryotic Phytoplankton (cells/ml)"]),
    "HNA Bacteria" : list(microbes_mean_coupeville.loc[:,"FACS High Nucleic Acid Bacteria (cells/ml)"]),
    "LNA Bacteria" : list(microbes_mean_coupeville.loc[:,"FACS Low Nucleic Acid Bacteria (cells/ml)"]),
    "Synechococcus" : list(microbes_mean_coupeville.loc[:,"FACS Synechococcus (cells/ml)"]),
    "Total Bacteria" : list(microbes_mean_coupeville.loc[:,"FACS Total Bacteria (cells/ml)"])
})

microbes_mean_coupeville_2 = pd.melt(microbes_mean_coupeville_1, id_vars = "Dates", var_name = "Microbes", value_name = "Concentration (cells per microliter)")
c = sns.catplot(data = microbes_mean_coupeville_2, x = "Dates", y = "Concentration (cells per microliter)", hue = "Microbes", kind = "bar")
c.set_xticklabels(rotation = 90)
plt.title("Average Cell Concentrations Collected Per Sample Day in Coupeville", loc = "center")
```

```python
microbes_mean_coupeville = coupeville.groupby(["Date"]).mean(numeric_only = True)
dates = ['08/24/2010', '08/25/2010', '08/26/2010', '11/21/2010', '11/23/2010', '02/15/2011', '02/16/2011', '02/17/2011', '05/17/2011', '05/18/2011', '05/19/2011']

microbes_mean_coupeville_1 = pd.DataFrame({
    "Dates" : dates,
    "Microbial Production (µg carbon/L /hour)" : list(microbes_mean_coupeville.loc[:,"Microbial Production (µg carbon/L /hour)"])
})

microbes_mean_coupeville_1.plot.line(x = "Dates", y = "Microbial Production (µg carbon/L /hour)", color = "#e5a37d")
plt.title("Average Microbial Production at Coupeville")
plt.legend(loc='upper center')
plt.ylabel("Microbial Production (µg carbon/L/hour)")
```
