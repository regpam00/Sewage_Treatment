DNA Fragments

```python
dna_frag = pd.read_excel("DNA Fragments (Cleaned).xlsx")
dna_frag.head(5)
```

```python
sns.set_theme(style="ticks")
f, ax = plt.subplots(figsize=(7, 5))
sns.despine(f)

sns.histplot(dna_frag, x = "Total Fragments", hue = "Site Name", multiple = "stack")
plt.title("Distribution of DNA Fragments from Collected Samples")
```

```python
sns.displot(dna_frag, x = "Total Fragments", hue = "Site Name", kde="True")
plt.title("Distribution of DNA Fragments from Collected Samples")
```
